```python
#http://toblerity.org/fiona/manual.html
import fiona
shp_file = "data/London-wards-2014_ESRI/London_Ward_CityMerged.shp"
c = fiona.open(shp_file, 'r')
```

```python
from fiona.crs import to_string
print to_string(c.crs)
c.crs
```

```python
import pyproj
from shapely.geometry import asShape, mapping
from shapely.ops import transform
dst_proj = pyproj.Proj('+init=epsg:4326')
src_proj = pyproj.Proj(to_string(c.crs))
wards = dict([(cc['properties']['NAME'], cc) for cc in c])
print asShape(wards['Highgate']['geometry']).is_valid
for ward in wards.values():
    ward['geometry'] = mapping(transform(lambda x, y, z=None: pyproj.transform(src_proj, dst_proj, x, y),
                                 asShape(ward['geometry'])))
    #ward['geometry'] = mapping(transform(lambda x, y, z=None: src_proj(x,y,inverse=True),
    #                             asShape(ward['geometry'])))

wards['Bunhill']
#wards['Highgate']
```

```python
import json
feature_collection = {"type": "FeatureCollection",
                      "features": wards.values()
                      }
with open('data/wards.json', 'w') as f:
    json.dump(feature_collection, f)
    
```

```python
import vincent
vincent.core.initialize_notebook()
wards_json = 'data/wards.topo.json'
geo_data = [{'name': 'wards',
             'url': wards_json,
              'feature': 'wards' 
            }]

vis = vincent.Map(geo_data=geo_data, scale=40000, center=[-0.0863623, 51.5209993], projection='mercator')
vis.display()

# topojson -s 0.000000001  -o wards.topo.small.json -- wards.json
wards_json = 'data/wards.topo.small.json'
geo_data = [{'name': 'wards',
             'url': wards_json,
              'feature': 'wards' 
            }]

vis = vincent.Map(geo_data=geo_data, scale=40000, center=[-0.0863623, 51.5209993], projection='mercator')
vis.display()

# topojson -s 0.00000001  -o wards.topo.extrasmall.json -- wards.json
wards_json = 'data/wards.topo.extrasmall.json'
geo_data = [{'name': 'wards',
             'url': wards_json,
              'feature': 'wards' 
            }]

vis = vincent.Map(geo_data=geo_data, scale=40000, center=[-0.0863623, 51.5209993], projection='mercator')
vis.display()

#
# 6.9M 19 Jun 10:12 wards.json
# 616K 19 Jun 10:12 wards.topo.json
# 98K 19 Jun 11:52 wards.topo.small.json
# 81K 19 Jun 11:54 wards.topo.extrasmall.json
```
